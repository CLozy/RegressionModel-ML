```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
automobile_df = pd.read_csv(r'C:\Users\Shaleen\Desktop\jpnb\datasets\cars_processed.csv')
automobile_df.head()
```

```python
automobile_df.shape
```

```python
X = automobile_df[['Age']]
Y = automobile_df['MPG']
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.scatter(automobile_df['Age'],automobile_df['MPG'])

plt.xlabel('Age')
plt.ylabel('MPG')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(normalize=True).fit(x_train, y_train)
```

```python
print('Training score:', linear_model.score(x_train, y_train))
```

```python
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('Testing score:', r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')

plt.xlabel('Age')
plt.ylabel('MPG')
plt.show()
```

```python
X =automobile_df[['Horsepower']]
Y = automobile_df['MPG']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

linear_model = LinearRegression(normalize=True).fit(x_train, y_train)

print('Training score:', linear_model.score(x_train, y_train))

y_pred = linear_model.predict(x_test)

print('Testing score:', r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')

plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```

```python
automobile_df = pd.get_dummies(automobile_df, columns=['Origin'])
automobile_df.head()
```

```python
X =automobile_df.drop('MPG', axis=1)
Y = automobile_df['MPG']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

linear_model = LinearRegression(normalize=True).fit(x_train, y_train)

print('Training score:', linear_model.score(x_train, y_train))

y_pred = linear_model.predict(x_test)

print('Testing score:', r2_score(y_test, y_pred))
```
